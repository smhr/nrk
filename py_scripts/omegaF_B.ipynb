```python
def fitMe (k, omega2):
    z = np.polyfit(k, omega2, 7) # Least squares polynomial fit
    p = np.poly1d(z) # Construct a polynomial with "z" coefficients
#     print np.poly1d(p)
    pDerive = p.deriv(1) # Construct 1st derivative coefficients
    pDerivePoly = np.poly1d(pDerive) # Construct 1st derivative polynomial
#     print pDerivePoly
    np.roots(pDerive) # Find roots of derivative
    for rootPD in np.nditer(np.roots(pDerive)):
        if (rootPD.imag == 0 and rootPD > 0.1 and rootPD < 0.5 ):
            kMin = rootPD.real
            print 'kMin, p(kMin) =', kMin, p(kMin)
            omeFast = math.sqrt(- p(kMin))
            print 'omeFast =', omeFast
    return kMin, p, omeFast
```
